```python
import numpy as np
import pandas as pd
```

Import the data from our github location

Lets set the patient id to int and change to category/object data types for our binary variables (scholarship, hipertension, etc.)

```python
#declare the categories
categ_features =['Scholarship','Hipertension','Diabetes','Alcoholism','Handcap','SMS_received']
#convert the data
df.PatientId = df.PatientId.astype(np.int64)
df[categ_features] = df[categ_features].astype(np.object)
```

Since we only have an appointment date and the schedule day only tells us the date time scheduled, we really don't need scheduled time of day... so we will convert both to date only

```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])

df['ScheduledDay'] =pd.DatetimeIndex(df.ScheduledDay).normalize()
```

### Cleaning the data

There are some spelling errors so we will correct those

```python
# Rename incorrect column names.
df = df.rename(columns={'Hipertension': 'Hypertension', 'Handcap': 'Handicap', 'SMS_received': 'SMSReceived', 'No-show': 'NoShow'})
```

Luckily its a pretty clean data set, but we do have a sub zero values for age and two unique patients with an age of 115. This could be documented oddly for perhaps a pregnant mother's child (hence a negative age value since the child isn't born). 

Since we don't know & its only one data point, we'll simply impute the value with the median for the negative. It is unlikely the 2 patients who are actually 115, but it is feasible to reach this age, and most likely, these patients true age are within the vicinity of the age documented but are slightly off due to poor birth documentation... so we will leave the 115 age values and other suspicously high age values alone for now.

```python
df.loc[df.Age==-1]
```

```python
df.loc[df.Age==115]
```

```python
df.Age=df.Age.replace(-1,37)
```

Next we will determine how many days in advance the appointment was scheduled and create it as a feature. We will then check and make sure values are reasonable and ensure there are no negative datapoints. Examination shows we do have some datapoints that need to be corrected

```python
df['DaysInAdvance']=(df['AppointmentDay']-df['ScheduledDay']).dt.days
df.loc[df.DaysInAdvance<0]
```

We will now run through our data and ensure we have no appointments that are scheduled after the appointment(which would be impossible). If true, we will replace the scheduled day with the appointment day assuming a same day appointment. Then we will recalculate our days in advance field

```python
df['ScheduledDay'] = np.where(df['ScheduledDay']>df['AppointmentDay'], df['AppointmentDay'], df['ScheduledDay'])
df['DaysInAdvance']=(df['AppointmentDay']-df['ScheduledDay']).dt.days
```

Examining again, we have corrected the bad scheduled appointment data

```python
df.loc[df.DaysInAdvance<0]
```

We will also create a day of week variable for both the scheduled day and the appointment day which will allows us to examine any potential trends related to the day of the week and appointment no-show

```python
df['ScheduledDOW'] = df['ScheduledDay'].dt.weekday_name
df['AppointmentDOW'] = df['AppointmentDay'].dt.weekday_name
```
